```python
import csv
import tkinter as tk
from tkinter import Canvas
```

```python
# =====================================
# MEMBACA DATASET
# =====================================

X = []
Y = []

with open("data.csv", "r") as file:

    reader = csv.DictReader(file)

    for row in reader:

        radius = float(row["radius_mean"])
        texture = float(row["texture_mean"])

        label = 1 if row["diagnosis"] == "M" else -1

        X.append([radius, texture])
        Y.append(label)

print("Jumlah data:", len(X))
```

```python
# =====================================
# NORMALISASI DATA
# =====================================

radius_values = [x[0] for x in X]
texture_values = [x[1] for x in X]

min_r = min(radius_values)
max_r = max(radius_values)

min_t = min(texture_values)
max_t = max(texture_values)

for i in range(len(X)):

    X[i][0] = (X[i][0] - min_r) / (max_r - min_r)

    X[i][1] = (X[i][1] - min_t) / (max_t - min_t)

print("Normalisasi selesai")
```

```python
# =====================================
# TRAIN TEST SPLIT
# =====================================

split_index = int(len(X) * 0.8)

X_train = X[:split_index]
Y_train = Y[:split_index]

X_test = X[split_index:]
Y_test = Y[split_index:]

print("Data Training :", len(X_train))
print("Data Testing  :", len(X_test))
```

```python
# =====================================
# TRAINING SVM MANUAL
# =====================================

w1 = 0.0
w2 = 0.0
b = 0.0

learning_rate = 0.01
lambda_param = 0.01
epochs = 500

for epoch in range(epochs):

    for i in range(len(X_train)):

        x1 = X_train[i][0]
        x2 = X_train[i][1]

        y = Y_train[i]

        condition = y * ((w1*x1)+(w2*x2)+b)

        if condition >= 1:

            w1 = w1 - learning_rate * (2 * lambda_param * w1)

            w2 = w2 - learning_rate * (2 * lambda_param * w2)

        else:

            w1 = w1 - learning_rate * (
                (2 * lambda_param * w1) - (y * x1)
            )

            w2 = w2 - learning_rate * (
                (2 * lambda_param * w2) - (y * x2)
            )

            b = b + learning_rate * y

print("Training selesai")
```

```python
# =====================================
# PREDIKSI
# =====================================

def predict(x1, x2):

    result = (w1*x1) + (w2*x2) + b

    if result >= 0:
        return 1

    return -1
```

```python
# =====================================
# CONFUSION MATRIX
# =====================================

TP = 0
TN = 0
FP = 0
FN = 0

for i in range(len(X_test)):

    actual = Y_test[i]

    pred = predict(
        X_test[i][0],
        X_test[i][1]
    )

    if actual == 1 and pred == 1:
        TP += 1

    elif actual == -1 and pred == -1:
        TN += 1

    elif actual == -1 and pred == 1:
        FP += 1

    elif actual == 1 and pred == -1:
        FN += 1

print("TP =", TP)
print("TN =", TN)
print("FP =", FP)
print("FN =", FN)
```

```python
# =====================================
# VALIDASI
# =====================================

accuracy = (TP + TN) / (TP + TN + FP + FN)

precision = TP / (TP + FP)

recall = TP / (TP + FN)

f1_score = 2 * (
    precision * recall
) / (
    precision + recall
)

print("Accuracy :", round(accuracy*100,2), "%")
print("Precision:", round(precision,4))
print("Recall   :", round(recall,4))
print("F1 Score :", round(f1_score,4))
```

```python
# =====================================
# VISUALISASI SVM
# =====================================

root = tk.Tk()

root.title("Support Vector Machine")
root.geometry("900x700")

title = tk.Label(
    root,
    text="Visualisasi SVM Breast Cancer",
    font=("Arial",16,"bold")
)

title.pack(pady=10)

canvas = Canvas(
    root,
    width=800,
    height=600,
    bg="white"
)

canvas.pack()

# =====================================
# GAMBAR TITIK DATA
# =====================================

for i in range(len(X)):

    x = 50 + (X[i][0] * 700)

    y = 550 - (X[i][1] * 500)

    if Y[i] == 1:
        color = "red"
    else:
        color = "blue"

    canvas.create_oval(
        x-3,
        y-3,
        x+3,
        y+3,
        fill=color,
        outline=color
    )

# =====================================
# DECISION BOUNDARY
# =====================================

if w2 != 0:

    x1 = 0
    y1 = -(w1*x1 + b)/w2

    x2 = 1
    y2 = -(w1*x2 + b)/w2

    screen_x1 = 50 + (x1 * 700)
    screen_y1 = 550 - (y1 * 500)

    screen_x2 = 50 + (x2 * 700)
    screen_y2 = 550 - (y2 * 500)

    canvas.create_line(
        screen_x1,
        screen_y1,
        screen_x2,
        screen_y2,
        fill="green",
        width=3
    )

# =====================================
# INFORMASI MODEL
# =====================================

info = tk.Label(
    root,
    text=f"w1 = {round(w1,4)} | w2 = {round(w2,4)} | b = {round(b,4)}",
    font=("Arial",12)
)

info.pack(pady=10)

root.mainloop()
```
